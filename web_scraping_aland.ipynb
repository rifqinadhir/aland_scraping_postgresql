```python
# Import library yang kita butuhkan
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
import time

# Inisialisasi DataFrame kosong untuk menyimpan hasil scraping
df = pd.DataFrame()

# Buat list kosong
productNames = []
brands = []
prices = []
productLinks = []

# Inisialisasi Selenium Chrome
driver = webdriver.Chrome()

# Menggunakan for loop untuk menelusuri 4 halaman dari situs web 
# Define target url yang ingin di tuju
# Memerintahkan browser untuk membuka halaman web
# Menambahkan delay 3 detik agar hasil screening lebih akurat
for page in range(1, 4):
    url = f"https://a-land.id/id/collections/trending-this-week?page={page}"
    driver.get(url)
    time.sleep(3)

    # Ambil HTML dari halaman yang dimuat oleh Selenium
    html = driver.page_source
    # Parsing HTML menggunakan BeautifulSoup
    soup = BeautifulSoup(html, 'html.parser')

    # Loop semua produk yang memiliki class "card-wrapper"
    for product in soup.find_all("div", class_="card-wrapper"):
        # Mencari elemen nama produk
        name = product.find("a", class_="full-unstyled-link")
        # Mencari elemen brand
        brand = product.find("a", class_="vendor-link")
        # Mencari elemen harga
        price = product.find("span", class_="price-item price-item--regular")
        
        # Menganmbil setiap elemen dari produk 
        # Jika elemen tidak ditemukan, maka di tulis none
        if name != None:
            productNames.append(name.get_text().strip())
        else:
            productNames.append(None)

        if brand != None:
            brands.append(brand.get_text().strip())
        else:
            brands.append(None)

        if price != None:
            prices.append(price.get_text().strip())
        else:
            prices.append(None)

        # Pertama, cari elemen <a> dengan class 'full-unstyled-link' yang biasanya berisi nama produk sekaligus link (name sudah didefinisikan sebelumnya)
        # Cek apakah elemen ditemukan dan memiliki atribut 'href'
        if name != None and name.has_attr("href"):
            # Ambil nilai href (misalnya: "/products/nama-produk")
            href = name["href"]
            # Gabungkan dengan domain utama untuk membentuk URL lengkap
            link = f"https://a-land.id{href}"
            # Simpan link ke dalam list
            productLinks.append(link)
        else:
            # Jika elemen tidak ditemukan atau tidak punya href, masukkan None
            productLinks.append(None)

# Tutup browser
driver.quit()

# Buat DataFrame
df['Nama Produk'] = productNames
df['Merek'] = brands
df['Harga'] = prices
df['Link Produk'] = productLinks

# Tampilkan semua baris
display(df)
```

```python
# Untuk menyimpan data ke file csv dengan nama "data_produk_aland"
df.to_csv('data_produk_aland.csv')
```
